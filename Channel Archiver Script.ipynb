```python
# import pytube
# https://github.com/GaziAdib/python-youtube-analyser-scrapper/blob/master/yt_channel.py

from pytube import Channel
from waybackpy import WaybackMachineSaveAPI
import time
import threading
```

```python
channel = Channel('https://www.youtube.com/channel/UC9K82e-1zNzYKRuZM0uheWA')
```

```python
def download_all_vids():
    start = time.time()
    progress_count = 0

    for video in channel.videos:
        try:
            video_time_start = time.time()
            video.streams.get_highest_resolution().download('seth yt backup')

            progress_count +=1 
            print(f'{progress_count}/{len(channel.videos)} ',video.title,f'has downloaded and took {round(time.time()-video_time_start,2)} seconds. \n')

        except:
            print(video.title,' failed to download :(\n')
            pass


    end = time.time()
    time_passed = (end-start)
    print(f'\n\nDownloading/archiving everything took this much time: {round(time_passed/60,2)} minutes.')

#     my_video.download('~/Downloads') https://stackoverflow.com/questions/65463820/change-download-location-to-downloads-in-pytube-in-python
```

```python
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:105.0) Gecko/20100101 Firefox/105.0'

def archive_all_vids():
    start = time.time()
    progress_count = 0

    for video in channel.videos:
        if progress_count == 0 or progress_count%15 == 0: 
            print('we need to wait before making more requests.')
            time.sleep(75)
            print('\nokay we waited 1.5 minutes!\n')
            progress_count=1

        else: 
            try:
                save_api = WaybackMachineSaveAPI(video.watch_url, user_agent)
                save_api.save()
                progress_count +=1 
                print(f'{video.title} successfully archived.')
            except:
                progress_count +=1 
                print(f'{video.title} failed to archive :(')
                pass 
    
    end = time.time()
    time_passed = (end-start)
    print(f'\n\n Archiving everything took this much time: {time_passed}')
    
archive_all_vids()
```

```python
if __name__ =="__main__":
    t1 = threading.Thread(target=download_all_vids)
    t2 = threading.Thread(target=archive_all_vids)
    
    t1.start()
    t2.start()
 
    t1.join()
    t2.join()
 
    # both threads completely executed
    print("\n\n\nBoth threads are done!")
```
